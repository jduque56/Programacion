Cordenadas polares

```python
import matplotlib.pyplot as plt
import numpy as np

theta2 = []
x=0
for i in range (0,13,1):
    theta2.append(x)
    x=x+np.pi/6
print(theta2)
r = [60,81,96,105,96,81,60,81,96,105,96,81,60]

fig = plt.figure()
ax = fig.add_subplot( projection="polar")
ax.plot(theta2,r)
print(r)
plt.show()
```
